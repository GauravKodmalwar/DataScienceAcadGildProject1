<h3>
12. Join users to transactions, keeping all rows from transactions and only matching rows from
users (left join)
</h3>

```python
# Identify common column names to join
print(transactions.columns)
print(transactions.shape)
```

```python
# Identify common column names to join
print(users.columns)
print(users.shape)
```

```python
# set index on common column
# join on common column after setting it as index
transaction_users_joined = transactions.set_index('UserID').join(users.set_index('UserID'), on='UserID', how='left')
transaction_users_joined
```

<h3>
21. Test to see if we can drop columns
</h3>

```python
transaction_users_joined
```

```python
nonMissingColumns = transaction_users_joined.dropna(thresh=int(transaction_users_joined.shape[0]*.9), axis=1).columns
nonMissingColumns
```

```python
missingColumns = transaction_users_joined.columns[transaction_users_joined.isnull().any() == True]
missingColumns
```

```python
for col in missingColumns:
    tmp = transaction_users_joined[transaction_users_joined[col].isnull()==True].shape[0]
    print('Number of missing values in {} column is {}'.format(col, tmp))
```

```python
totalRows = transaction_users_joined.shape[0]

for col in missingColumns:
    tmp = transaction_users_joined[transaction_users_joined[col].isnull()==True].shape[0]
    print('Percent of missing values in {} column is {}'.format(col, tmp/totalRows))
```

```python
# To drop columns, just get the required columns in the output dataframe
transaction_users_joined[nonMissingColumns]
```
